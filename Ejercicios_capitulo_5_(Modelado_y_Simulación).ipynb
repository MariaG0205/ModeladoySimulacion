##**5.1.1**
(Elipses y conservación de energía para el oscilador armónico) Considere
el oscilador armónico $\dot{x}= v$, $\dot{v}= −  \omega^2x$.

**a)** Demuestre que las órbitas están dadas por elipses $\omega^2x^2 + v^2  =  C$, donde $C$ es cualquier constante no negativa. (Sugerencia: divida la ecuación $\dot{x}$   por la ecuación $\dot{v}$  , separe las $v$ de las $x$ e integre la ecuación separable resultante.)


```python
def vdp(t,z):
  x,y=z 
  #Donde, v=y
  return[y,-mu**2*x] #Ecuaciones del sistema

a,b=0,20
mus=[0,1,3,5]
styles=["-","--",":","_"]
t=np.linspace(a,b,500)

for mu,style in zip(mus,styles):
  sol=solve_ivp(vdp,[a,b],[1,0],t_eval=t)
  plt.plot(sol.y[0],sol.y[1],style)
plt.legend([f"$\mu={m}$" for m in mus])
plt.grid()
```
