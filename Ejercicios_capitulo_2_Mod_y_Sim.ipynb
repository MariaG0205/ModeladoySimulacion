```
def step_euler(y,h,f):
  next_y=y+h*f(y)
  return next_y

def euler(h,f,y_0,start_t,end_t):
  N= int((end_t-start_t)/h)
  t_list=np.linspace(start_t, end_t, N+1)

  y_list=np.zeros(N+1)
  y_list[0]=y_0
  
  for i in range(0,N):
    y_list[i+1]=step_euler(y_list[i],h,f)
  return t_list,y_list
```

```
def g(y):
  ec = np.sin(y)
  return ec

x= np.linspace(-2*np.pi,2*np.pi,25)
y_0=[0, np.pi/4, np.pi/2]
h=0.01
t_0=0
t_N=10


for y in x:
  t_list, y_list= euler(h,g,y,t_0,t_N)
  plt.plot(t_list,y_list)
  plt.grid()
  plt.ylabel('X(t)')
  plt.xlabel('t')
```

**c)** Grafique el error $E = \hat{x}(1)−x(1)$ en función de  $\triangle t$. Luego grafique $ln E$ contra $ln t$. Explica los resultados.
